```python
import torch
from PPO.PPO import PPO
from PPO.PPO import run_test_env
from PPO.wrappers import make_atari, CropScreen, wrap_deepmind, wrap_pytorch, ExcludeFireAction
```

```python
# HYPER-PARAMETERS #
max_frames       = int(1e7)
max_rewards      = None
num_steps        = 128
num_envs         = 8
batch_size       = 256
ppo_epochs       = 4
lr               = 'lin_2.5e-4'
clip_range       = 0.1
vf_coef          = 0.5
ent_coef         = 0.01
gamma            = 0.99
lam              = 0.7
clip_grad_norm   = 0.5
std_decay        = True
clip_log_std     = None
log_every        = 100
runs_for_episode = 5
test_episodes    = 5
log_file         = './logs/Breakout-v4.pth'
```

```python
def make_env():
    def _thunk():
        env = make_atari('BreakoutNoFrameskip-v4')
        env = CropScreen(env, top=32, bottom=225, left=8, right=160)
        env = wrap_deepmind(env, episode_life=True, clip_rewards=True, frame_stack=True, scale=True)
        env = ExcludeFireAction(env)
        env = wrap_pytorch(env)
        return env
    return _thunk

def make_test_env():
    env = make_atari('BreakoutNoFrameskip-v4')
    env = CropScreen(env, top=32, bottom=225, left=8, right=160)
    env = wrap_deepmind(env, episode_life=True, clip_rewards=False, frame_stack=True, scale=True)
    env = ExcludeFireAction(env)
    env = wrap_pytorch(env)
    return env
```

```python
import matplotlib.pyplot as plt

env = make_env()()
print('obs_space:', env.observation_space)
print('act_space:', env.action_space)
obs = env.reset()
for i in range(10):
    obs, r, d, _ = env.step(torch.randint(0, 3, (1,)).item())
fig, axs = plt.subplots(1, 4)
axs[0].imshow(obs[0], cmap='gray')
axs[1].imshow(obs[1], cmap='gray')
axs[2].imshow(obs[2], cmap='gray')
axs[3].imshow(obs[3], cmap='gray')
```

```python
ppo = PPO(make_env, device='auto')
```

```python
# Run this cell to load a previous checkpoint
ppo.load(log_file)
```

```python
ppo.learn(max_frames, max_rewards=max_rewards, num_steps=num_steps, num_envs=num_envs, batch_size=batch_size,
          ppo_epochs=ppo_epochs, clip_range=clip_range, lr=lr, vf_coef=vf_coef, ent_coef=ent_coef,
          gamma=gamma, lam=lam, clip_grad_norm=clip_grad_norm, std_decay=std_decay, clip_log_std=clip_log_std,
          log_every=log_every, log_file=log_file, test_env=make_test_env(), runs_for_episode=runs_for_episode, test_episodes=test_episodes)
```

```python
del ppo
```

```python
ppo = PPO(make_env, device='cpu')
```

```python
ppo.load(log_file)
```

```python
env = make_test_env()
runs_for_episode = 5
n_episode = 10

mean_reward = run_test_env(env=env, model=ppo, runs_for_episode=runs_for_episode, n_episodes=n_episode, vis=True, deterministic=True)
env.close()

print('Mean reward over %i episodes = %f' % (n_episode, mean_reward))
```
