```python
from PPO.PPO import PPO
from PPO.PPO import run_test_env
from PPO.wrappers import make_atari, wrap_deepmind, wrap_pytorch
```

```python
# HYPER-PARAMETERS #
max_frames  = int(1e7)
max_rewards = None
num_steps   = 128
num_envs    = 8
batch_size  = 128
ppo_epochs  = 4
clip_range  = 'lin_0.15'
lr          = 'lin_3e-4'
vf_coef     = 0.5
ent_coef    = 0.01
gamma       = 0.99
tau         = 0.8
clip_grad   = 0.5
log_every   = 50
test_runs   = 5
log_file    = './logs/Breakout-v4.pth'
```

```python
def make_env():
    def _thunk():
        env = make_atari('BreakoutNoFrameskip-v4')
        env = wrap_deepmind(env, episode_life=True, clip_rewards=False, frame_stack=True, scale=True)
        env = wrap_pytorch(env)
        return env
    return _thunk

def make_test_env():
    env = make_atari('BreakoutNoFrameskip-v4')
    env = wrap_deepmind(env, episode_life=False, clip_rewards=False, frame_stack=True, scale=True)
    env = wrap_pytorch(env)
    return env
```

```python
ppo = PPO(make_env, device='auto')
```

```python
# Run this cell to load a previous checkpoint
ppo.load(log_file)
```

```python
ppo.learn(max_frames, max_rewards=max_rewards, num_steps=num_steps, num_envs=num_envs, batch_size=batch_size,
          ppo_epochs=ppo_epochs, clip_range=clip_range, lr=lr, vf_coef=vf_coef, ent_coef=ent_coef, gamma=gamma, tau=tau, clip_grad_norm=clip_grad,
          log_every=log_every, plot=True, save=True, log_file=log_file, test_env=make_test_env(), test_runs=test_runs)
```

```python
del ppo
```

```python
ppo = PPO(make_env, device='cpu')
```

```python
ppo.load(log_file)
```

```python
env = make_test_env()
n_episode = 3

mean_reward = run_test_env(env=env, model=ppo, n_runs=n_episode, vis=True, deterministic=False)
env.close()

print('Mean reward over %i episodes = %f' % (n_episode, mean_reward))
```
